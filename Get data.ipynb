```python
def getAnimesOfSeason(season, year):
    response = requests.get('https://api.jikan.moe/v3/season/{}/{}'.format(year, season))
    return response.json()['anime']
```

```python
getAnimesOfSeason('spring', 2020)
```

```python
getAnimesOfSeason('fall', 2020)
```

```python
getAnimesOfSeason('winter', 2020)
```

```python
animes = getAnimesOfSeason('fall', 1999)
for anime in animes:
    print(anime['title'], 
             anime['url'], 
             anime['synopsis'].replace('\n[Written by MAL Rewrite]', '').strip().rstrip('\n'),
             ', '.join([genre['name'] for genre in anime['genres']]),
          anime['score']
            )
```

```python
from time import sleep
```

```python
allAnimes = []
for year in range(2000,2021):
    for season in ['spring', 'summer', 'fall', 'winter']:
        animes = getAnimesOfSeason(season, year)
        allAnimes.extend(animes)
        print('Got {} animes for {} {}'.format(len(animes), season, year))
        sleep(5)

with open('animes.csv', mode='w', encoding='utf-8', newline='') as f:
    anime_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    anime_writer.writerow(['title', 'URL', 'synopsis', 'genre', 'score'])
    for anime in allAnimes:
        anime_writer.writerow(
            [anime['title'], 
             anime['url'], 
             anime['synopsis'].replace('\n[Written by MAL Rewrite]', '').strip().rstrip('\n'),
             ', '.join([genre['name'] for genre in anime['genres']]),
             anime['score']
            ])
        
print(len(allAnimes))
```
